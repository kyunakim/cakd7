```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
from keras import Model
from keras.models import Sequential
from keras.layers import Dense, BatchNormalization, Dropout, Conv2D, MaxPooling2D, Flatten, Reshape, Input
from keras.callbacks import ReduceLROnPlateau, EarlyStopping, ModelCheckpoint
from keras.utils import to_categorical
from keras.datasets.fashion_mnist import load_data
from keras.preprocessing.image import ImageDataGenerator
```

```python
(x_train, y_train),(x_test,y_test) = load_data()
x_train = x_train[...,np.newaxis]
x_test = x_test[..., np.newaxis]

x_train = x_train/255.
x_test = x_test/255.

print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
class_names = ['T-Shirt','Trouser','Pullover','Dress','Coat','Sandal','Shirt','Sneaker','Bag','Ankle Boot']
```

```python
# 모델 구성 및 컴파일

def build_model():
  model = Sequential()
  input = Input(shape=(28,28,1))
  output = Conv2D(filters=32, kernel_size=(3,3))(input)
  output = Conv2D(filters=64, kernel_size=(3,3))(output)
  output = Conv2D(filters=64, kernel_size=(3,3))(output)
  output = Flatten()(output)
  output = Dense(128,activation='relu')(output)
  output = Dense(64,activation='relu')(output)
  output = Dense(10,activation='softmax')(output)
  
  model = Model(inputs=[input],outputs=output)
  model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['acc'])
  return model

model_1 = build_model()
model_1.summary()
```

```python
hist_1 = model_1.fit(x_train,y_train,epochs=25,validation_split=0.3,batch_size=128)
```

```python
hist_1.history.keys()
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(hist_1.history['loss'],'b--',label='loss')
plt.plot(hist_1.history['val_loss'],'r',label='val_loss')
plt.xlabel('Epochs')
plt.grid()
plt.legend()

plt.subplot(1,2,2)
plt.plot(hist_1.history['acc'],'b--',label='accuracy')
plt.plot(hist_1.history['val_acc'],'r',label='val_acc')
plt.xlabel('Epochs')
plt.grid()
plt.legend()
```

```python
model_1.evaluate(x_test,y_test)
```

### 모델 구성 및 학습 : 과대적합 방지
- 슬림한 모델로 변경

```python
def build_model2():
  model = Sequential()
  input = Input(shape=(28,28,1))
  output = Conv2D(filters=16, kernel_size=(3,3))(input)
  output = MaxPooling2D(strides=(2, 2))(output)
  output = Conv2D(filters=32, kernel_size=(3,3))(output)
  output = Dropout(0.4)(output)
  output = MaxPooling2D(strides=(2, 2))(output)
  output = Conv2D(filters=32, kernel_size=(3,3))(output)
  output = Dropout(0.4)(output)
  output = MaxPooling2D(strides=(2, 2))(output)  
  output = Flatten()(output)
  output = Dense(64,activation='relu')(output)
  output = Dropout(0.5)(output)
  output = Dense(32,activation='relu')(output)
  output = Dropout(0.5)(output)
  output = Dense(10,activation='softmax')(output)
  
  model = Model(inputs=[input],outputs=output)
  model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['acc'])
  return model

model_2 = build_model2()
model_2.summary()
```

```python
hist_2 = model_2.fit(x_train,y_train,epochs=25,validation_split=0.3,batch_size=128)
```

```python
hist_2.history.keys()
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(hist_2.history['loss'],'b--',label='loss')
plt.plot(hist_2.history['val_loss'],'r',label='val_loss')
plt.xlabel('Epochs')
plt.grid()
plt.legend()

plt.subplot(1,2,2)
plt.plot(hist_2.history['acc'],'b--',label='accuracy')
plt.plot(hist_2.history['val_acc'],'r',label='val_acc')
plt.xlabel('Epochs')
plt.grid()
plt.legend()
```

```python
model_2.evaluate(x_test,y_test)
```

```python
from keras.layers import BatchNormalization
```

```python
def build_model3():
  model = Sequential()
  input = Input(shape=(28,28,1))

  output = Conv2D(filters=64, kernel_size=3, activation='relu',padding='same')(input)
  output = BatchNormalization()(output)
  output = Conv2D(filters=64, kernel_size=3, activation='relu',padding='valid')(output)
  output = Dropout(0.5)(output)
  output = MaxPooling2D(strides=(2, 2))(output)

  output = Conv2D(filters=128, kernel_size=3, activation='relu',padding='valid')(output)
  output = BatchNormalization()(output)
  output = Conv2D(filters=128, kernel_size=3, activation='relu',padding='valid')(output)
  output = Dropout(0.5)(output)
  output = MaxPooling2D(strides=(2, 2))(output)


  output = Flatten()(output)

  output = Dense(512,activation='relu')(output)
  output = Dropout(0.5)(output)

  output = Dense(256,activation='relu')(output)
  output = Dropout(0.5)(output)

  output = Dense(10,activation='softmax')(output)
  
  model = Model(inputs=[input],outputs=output)
  model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['acc'])
  return model

model_3 = build_model3()
model_3.summary()
```

```python
hist_3 = model_3.fit(x_train,y_train,epochs=25,validation_split=0.3,batch_size=128)
```

```python
hist_3.history.keys()
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(hist_3.history['loss'],'b--',label='loss')
plt.plot(hist_3.history['val_loss'],'r',label='val_loss')
plt.xlabel('Epochs')
plt.grid()
plt.legend()

plt.subplot(1,2,2)
plt.plot(hist_3.history['acc'],'b--',label='accuracy')
plt.plot(hist_3.history['val_acc'],'r',label='val_acc')
plt.xlabel('Epochs')
plt.grid()
plt.legend()
```

### 모델 성능 높이기 : Image Augmentation

```python
from keras.layers.preprocessing import image_preprocessing
from keras.preprocessing.image import ImageDataGenerator

image_generator = ImageDataGenerator(
    rotation_range= 10,
    zoom_range = 0.2,
    shear_range= 0.6,
    width_shift_range= 0.1,
    height_shift_range=0.1,
    horizontal_flip=True,
    vertical_flip=False
)

augument_size = 200
print(x_train.shape)
print(x_train[0].shape)
```

```python
x_agument = image_generator.flow(np.tile(x_train[0].reshape(28*28*1),augument_size).reshape(-1,28,28,1),
                                         np.zeros(augument_size), batch_size=augument_size,shuffle=False).next()[0]
```

```python
plt.figure(figsize=(10,10))
for i in range(1,101):
  plt.subplot(10,10,i)
  plt.axis('off')
  plt.imshow(x_agument[i-1].reshape(28,28),cmap='gray')
```

```python
image_generator = ImageDataGenerator(
    rotation_range= 15,
    zoom_range = 0.1,
    shear_range= 0.6,
    width_shift_range= 0.15,
    height_shift_range=0.1,
    horizontal_flip=True,
    vertical_flip=False
)

augment_size = 30000
random_mask = np.random.randint(x_train.shape[0],size=augment_size)
x_agumentend = x_train[random_mask].copy()
y_agumentend = y_train[random_mask].copy()
```

```python
x_agumentend = image_generator.flow(x_agumentend,np.zeros(augment_size),
                                    batch_size=augment_size,shuffle=False).next()[0]
```

```python
x_train = np.concatenate((x_train,x_agumentend))
y_train = np.concatenate((y_train,y_agumentend))
print(x_train.shape)
print(y_train.shape)
```

```python
model_4 = build_model3()
model_4.summary()
```

```python
hist_4 = model_4.fit(x_train,y_train,epochs=40,validation_split=0.3,batch_size=128)
```

```python
hist_4.history.keys()
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(hist_4.history['loss'],'b--',label='loss')
plt.plot(hist_4.history['val_loss'],'r',label='val_loss')
plt.xlabel('Epochs')
plt.grid()
plt.legend()

plt.subplot(1,2,2)
plt.plot(hist_4.history['acc'],'b--',label='accuracy')
plt.plot(hist_4.history['val_acc'],'r',label='val_acc')
plt.xlabel('Epochs')
plt.grid()
plt.legend()
```

```python
model_4.evaluate(x_test,y_test)
```
