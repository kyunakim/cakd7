## 그룹화 (groupping)

### 그룹 연산
- 복잡한 데이터를 어떤 기준에 따라 여러 그룹으로 나누어서 관찰할 수 있으며 이런 방식으로 분할 처리하는 것을 그룹 연산이라 함.
- 그룹 연산은 데이터를 집계, 반환, 필터하는데 효율적이며 판다스 groupby() 메소드를 사용
- 그룹 객체 만들기(분할)
- 그룹 연산 메소드(적용-결합)

```python
import pandas as pd
import seaborn as sns #seaborn 라이브러리 불러오기 (dataset)

titanic = sns.load_dataset('titanic')
titanic.columns
```

```python
titanic.head()
```

```python
import pandas as pd
import seaborn as sns

titanic = sns.load_dataset('titanic')
df = titanic.loc[:,['age','sex','class','fare','survived']]

print('승객 수:', len(df))
print(df.head())
```

```python
df.columns
```

#### 그룹 객체 만들기 (분할) group by

```python
# class열을 기준으로 그룹핑(분할)
grouped = df.groupby(['class'])
print(grouped)
list(grouped)
```

```python
# 그룹 객체를 iteration으로 출력
for key, group in grouped:
    print('* key :', key)
    print('* number :',len(group))
    print(group.head())
```

#### groupping  함수

```python
# 연산 메서드 적용
average = grouped.mean() # mean 평균값 구하기 
average
```

```python
stat = grouped.max() # max 최댓값 구하기
stat
```

```python
# Q. 3rd 그룹만을 선택해서 group3 이름으로 저장하고 통계요약표를 출력하세요.
group3 = grouped.get_group('Third') # 3rd 그룹 group3로 저장하기
group3.describe() # describe 각 column별로 다양한 정보 한번에 보기
```

```python
# 2개의 column지정해서 groupby하기

grouped_two = df.groupby(['class','sex'])
average_two = grouped_two.mean()
average_two
```

```python
#  get_group 그룹화한 해당 행만 선택하여 보기 / 특정 데이터 추출
# get_group() : 기준으로 한 열의 값을 가진 그룹만을 추출해 낼 수 있다.
group3f = grouped_two.get_group(('Third','female'))
group3f.head()
```

#### 그룹 연산 메소드 (적용-결합)

```python
# 모든 열에 여러 함수를 매핑
agg_all = grouped.agg(['min','max'])
agg_all.head()
```

```python
# 각 열마다 다른 함수를 적용하여 집계
# 그룹 객체에 함수를 매핑할 경우엔 agg()메소드를 사용한다.
agg_sep = grouped.agg({'fare':['min','max'],'age':'mean'})
agg_sep.head()
```

### 피벗 테이블 (pivot)
피벗 테이블을 구성하는 4가지 요소 (행인덱스, 열인덱스, 데이터 값, 데이터 집계함수)에 적용할 데이터프레임의 열을 각각 지정하여 함수의 인자로 전달

```python
import pandas as pd
import seaborn as sns
titanic = sns.load_dataset('titanic')
df = titanic.loc[:,['age','sex','class','fare','survived']]
df.head()
```

```python
pdf1 = pd.pivot_table(df,
                     index = 'class',  # 행인덱스
                     columns = 'sex',   # 열인덱스
                     values = 'fare',   # 데이터 값
                      aggfunc = 'mean')   #데이터 집계함수
pdf1.head()
```

```python
# 행, 열, 값에 사용할 열을 2개 이상 지정 - index, values, aggfunc에 각각 2개 이상 지정
pdf2 = pd.pivot_table(df,
                     index=['class','sex'],
                     columns='survived',
                     values=['age','fare'],
                     aggfunc=['mean','max'])
pdf2.head()
```

[pipe vs applymap vs apply]  

pipe와 applymap과의 차이는 반환하는 결과가 pipe가 더 다양하다는 것으로 applymap은 데이터프레임 원소에 적용하므로 데이터프레임으로만 리턴되지만 pipe는 사용자 함수의 리턴값에 따라 개별 값, 시리즈, 데이터프레임을 반환할 수 있다.  

현재 가장 많이 쓰이는 메서드는 apply인데 그 이유는 다양한 형태로 값을 입력 받을 수 있고, 반환하는 형태도 폭이 넓기 때문. 결과는 같아보이지만, 시리즈를 이용해 데이터프레임의 행과 열도 이용할 수 있으며 시리즈 그 자체도 함수로 활용할 수 있다. 

만약 데이터프레임에 함수를 적용해야한다면 리턴하는 형태가 많은 pipe 사용도 고려


```python
import pandas as pd
df = pd.DataFrame({
    'col1':[30,40,50,60],
    'col2':[23,35,65,45],
    'col3':[85,87,90,89],   
}, index = list('ABCD'))
df
```

```python
df.apply(lambda x:x.sum(),axis=0)
```

```python
df.col1.apply(lambda x:x+1)
```

```python
df.col1.map(lambda x : x+1)
```

```python
df.applymap(lambda x:'%.2f'% x)
```

```python
# The pipe() method allows you to apply one or more functions to the DataFrame object.

import pandas as pd

def change_age(x):
  x["age"]=[10, 20, 30]
  return x

data = {
  "name": ["Sally", "Mary", "John"],
  "age": [50, 40, 30]
}

df = pd.DataFrame(data)
df
```

```python
df.pipe(change_age)
```

Q14. 각 열의 NaN 찾기, 각 열의 NaN 개수 반환, 데이터프레임의 총 NaN 개수를 반환하는 사용자 함수를 정의한 후 df.pipe 메소드로 함수 매핑한 결과를 모두 출력하세요.



```python
import pandas as pd
import seaborn as sns

titanic = sns.load_dataset('titanic')

# 각 열의 NaN 찾기
def missing_value(x):
    return x.isnull()

# 각 열의 NaN 개수 반환
def missing_count(x):
    return missing_value(x).sum()

# 데이터프레임의 총 NaN 개수를 반환
def total_number_missing(x):
    return missing_count(x).sum()
```

```python
result = titanic.pipe(missing_value)
result
```

```python
result2 = titanic.pipe(missing_count)
result2
```

```python
result3 = titanic.pipe(total_number_missing)
result3
```

```python
# Q. horsepower를 3개의 bin으로 나누고 bin_names = ['저출력','보통출력','고출력']으로 출력하세요.
# cut 함수 : 실수 값의 경계를 지정하는 함수. 각 범주가 모두 동일한 길이를 갖도록 데이터를 범주화


import numpy as np
count, bin_dividers = np.histogram(auto_df.horsepower,bins=3) 
print(bin_dividers)
bin_names = ['저출력','보통출력','고출력']
auto_df['hp_bin'] = pd.cut(x=auto_df.horsepower, # 데이터 배열 
                          bins=bin_dividers,     # 경계값 리스트
                          labels = bin_names,    # bin 이름
                          include_lowest=True) # 첫 경계깞 포함

auto_df[['horsepower','hp_bin']].head()
```
