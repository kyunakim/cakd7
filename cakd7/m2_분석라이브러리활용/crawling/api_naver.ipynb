```python
import os
import sys
import urllib.request

client_id = "OHjow4RoV5IW9hbt4e6H"
client_secret = "WX9_Vqfv6p"

query = urllib.parse.quote(input("검색 질의 "))
url = 'https://openapi.naver.com/v1/search/webkr.json?query='+query # json 결과
# url = "https://openapi.naver.com/v1/search/webkr.json?query=" +qurey # xml 결과


request = urllib.request.Request(url)
request.add_header("X-Naver-Client-Id",client_id)
request.add_header("X-Naver-Client-Secret",client_secret)
response = urllib.request.urlopen(request)
rescode = response.getcode()

import json
if(rescode==200):
    response_body = response.read()
    data = json.loads(response_body)
    #print(response_body.decode('utf-8'))
else:
    print("Error Code:" + rescode)

import warnings
warnings.filterwarnings('ignore')
import pandas as pd

length=len(data['items'])

title=[]
link=[]
description=[]
for i in range(length):
    title.append(data['items'][i]["title"])
    description.append(data['items'][i]["description"])
    link.append(data['items'][i]["link"])



df = pd.DataFrame({"title":title,"description":description,"link":link})
#df = pd.read_json(response_body.decode('utf-8'))

import re
for i in range(10):
    df[title] = re.sub('[<br></br>]')
```
