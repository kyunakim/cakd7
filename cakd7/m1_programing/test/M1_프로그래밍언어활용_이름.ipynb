#### 2. 아래 사용자 함수, 변수 리스트 및 2개의 내장함수를 각각 이용해서  [1, 4, 9, 16, 25], [1, 2]를 출력하세요.(내장함수 2개 사용 필수) 

```python
# 함수를 선언합니다.
def power(item):
    return item * item
def under_3(item):
    return item < 3

# 변수를 선언합니다.
list_input_a = [1, 2, 3, 4, 5]
```

#### 11. 정규표현식을 사용하여 text에서 지역코드만 출력하세요.
text = "문의사항이 있으면 032-232-3245 으로 연락주시기 바랍니다."

```python
import re
text = "문의사항이 있으면 032-232-3245 으로 연락주시기 바랍니다."
p = re.compile(r"\w+\s+(\d{3})+[-]\w+")
m = p.finditer(text)
for i in m:
    print(i.group(1))  
```

#### 12. 'abcdefghij' 에 대하여 중첩을 적용한 서브그룹 5개로 컴파일하여 group()함수를 이용하여 'abcdefghi'와 e'를 출력하세요

```python
import re
data = 'abcdefghij'
p = re.compile("(\w{2}\w{2}(\w{1})\w{3}\w{2})")
m = p.finditer(data)
for i in m:
    print(i.group(0))
    print(i.group(2))
```

####  16. numbers 내부에 들어 있는 숫자가 몇 번 등장하는지를 출력하는 코드를 작성하세요.

```python
numbers = [3,6,3,5,8,2,9,7,4,3,6,3,5,8,2,9,7,4]
count = {}
for number in numbers:
    if number in count:count[number]+=1
    else : count[number]=1
print(count)
```

```python
numbers = [3,6,3,5,8,2,9,7,4,3,6,3,5,8,2,9,7,4]
count = {}
for number in numbers:                       # number 변수에 numbers를 넣기
    if number in count:                     # number가 counter에 있는지 확인
        count[number] += 1                # 있는 경우엔 +1 해줌
    else:
        count[number] = 1                 # 없는 경우엔 1로 초기화 해줌
print(count)
```

#### 18. 사칙연산을 수행하는  클래스를 작성하여 객체를 만들고 2개의 정수 10, 5를 입력한 후 연산 결과를  포멧을 이용하여 출력하세요.

```python
class FourCal:
    def __init__(self,first,second):
        self.first = first
        self.second = second
    def add(self):
        result = self.first + self.second
        return result
    def subtract(self):
        result = self.first - self.second
        return result
    def multiply(self):
        result = self.first * self.second
        return result
    def divide(self):
        result = self.first / self.second
        return result
```

```python
num1 = int(input("숫자"))
num2 = int(input("숫자"))
```

```python
a = FourCal(num1,num2)
```

```python
print(f"덧셈값은 {a.add()}, 뺄셈값은 {a.subtract()}, 곱셈값은 {a.multiply()}, 나눗셈값은 {a.divide()}이다.")
```
